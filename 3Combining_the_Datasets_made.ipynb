```python
import pandas as pd

df = pd.read_csv("/content/telugu_political_news2.csv")

df["article"] = df["title"].astype(str) + " " + df["description"].astype(str)

df = df[["article", "party"]]

df.to_csv("Db2.csv", index=False)

print("Cleaned file saved as 'telugu_political_articles.csv'")
print(df.head())
```
